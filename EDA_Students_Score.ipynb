#Gender Distributioon


```python
df = pd.DataFrame(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize= (5,5))
ax = sns.countplot(data=df, x='Gender', palette={'male': 'skyblue', 'female': 'pink'})
plt.title("Gender Distribution")
for container in ax.containers:
    ax.bar_label(container)

plt.show()
```

We have analyzed from this chart that no. of count of female is greater than the male in the dateset

```python
df.head()
```

```python
# Finding the effect of Parents education on students to theri respective subject marks

gb = df.groupby(['ParentEduc', 'ParentMaritalStatus']).agg({'MathScore':'mean', 'ReadingScore':'mean', 'WritingScore':'mean'})
```

```python
print(gb)
```

From this we can say Divroced parents marks is higher in every degree

```python
gb_2 = df.groupby('ParentEduc').agg({'MathScore':'mean', 'ReadingScore':'mean', 'WritingScore':'mean'})
```

```python
print(gb_2)
```

```python
plt.figure(figsize= (5,5))
sns.heatmap(gb_2, annot = True)
plt.title("Parents's Education vs marks")
```

From the chart we have concluded that whose parents have good education their children also have good academic marks
